```python
import pandas as pd

path = '/home/abdul/sales-analytics-dashboard/data/raw/Sample - Superstore.csv'
df = pd.read_csv(path, encoding='latin-1')
print("Shape:", df.shape)
```

```python
# Total nulls per column
print("=== Missing Values ===")
print(df.isnull().sum())

print("\n=== Missing Percentage ===")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
print("Total rows:", len(df))
print("Duplicate rows:", df.duplicated().sum())

# If any duplicates found, remove them
df = df.drop_duplicates()
print("Rows after deduplication:", len(df))
```

```python
# Fix Order Date and Ship Date
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date']  = pd.to_datetime(df['Ship Date'])

# Fix Postal Code â must be string to preserve leading zeros
df['Postal Code'] = df['Postal Code'].astype(str)

# Confirm the fixes
print(df[['Order Date', 'Ship Date', 'Postal Code']].dtypes)
print("\nSample dates:")
print(df[['Order Date', 'Ship Date']].head(3))
```

```python
#Feature Engineering
# How many days between order placed and order shipped?
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days

print("Shipping Days stats:")
print(df['Shipping Days'].describe())
```

```python
#Time column for trend analysis
df['Year']       = df['Order Date'].dt.year
df['Month']      = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%b')   # Jan, Feb, Mar...
df['YearMonth']  = df['Order Date'].dt.to_period('M').astype(str)  # 2014-01

print("Unique years in dataset:", sorted(df['Year'].unique()))
print("Sample YearMonth values:", df['YearMonth'].head(3).tolist())
```

```python
#Profit margin 
# Profit as a percentage of Sales
df['Profit Margin %'] = (df['Profit'] / df['Sales'] * 100).round(2)

print("Profit Margin % stats:")
print(df['Profit Margin %'].describe())
```

```python
#Revenue tier(segmentation)
df['Revenue Tier'] = pd.cut(
    df['Sales'],
    bins=[0, 100, 500, 1000, 99999],
    labels=['Low', 'Medium', 'High', 'Very High']
)

print("\nRevenue Tier distribution:")
print(df['Revenue Tier'].value_counts())
```

```python
#verifying the cleaned dataset 
print("=== Final Shape ===")
print(df.shape)

print("\n=== All Columns ===")
print(df.columns.tolist())

print("\n=== Data Types ===")
print(df.dtypes)

print("\n=== Sample of new columns ===")
df[['Order Date', 'Shipping Days', 'Year', 'YearMonth', 'Profit Margin %', 'Revenue Tier']].head()
```
